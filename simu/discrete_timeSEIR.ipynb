```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# SEIR model and mean field approximation 

```python
alphaE=0.01
alphaI=0.08
alphaR=0.02
alphaL=0.01
alphaA=0.04
def computeK(m):
    K = np.zeros(shape=(4,4))
    K[0,1] = alphaE+alphaI*m[2];     K[0,0] = 1-K[0,1]
    K[1,2] = alphaA;                 K[1,1] = 1-K[1,2]
    K[2,3] = alphaR;                 K[2,2] = 1-K[2,3]
    K[3,0] = alphaL;                 K[3,3] = 1-K[3,0]
    for i in range(3):
        for j in range(3):
            if K[i,j] < 0 or K[i,j] > 1: 
                print('rate error: K[',i,j,']=',K[i,j])
    return(K)

m0 = np.array([1,0,0,0])

def simulate(m0,T):
    m=m0
    M=np.zeros(shape=(T,4))
    for t in range(T):
        m = np.tensordot(m,computeK(m),1)
        M[t,:] = m
    return(M)
```

```python
M = simulate(m0,500)
plt.plot(M[:,0])
plt.plot(M[:,1])
plt.legend(['S','I'])
```

# Refined mean field approximation 

```python
def computeA(m):
    return(np.array([
        [1-(alphaE+alphaI*m[2]), 0,          -alphaI*m[0],  alphaL],
        [alphaE+alphaI*m[2],     1-alphaA,   alphaI*m[0],   0],
        [0,                      alphaA,     1-alphaR,      0],
        [0,                      0,          alphaR,        1-alphaL]]))
def computeB(m):
    B = np.zeros(shape=(4,4,4))
    B[0,0,2] = -alphaI
    B[0,2,0] = -alphaI
    B[1,0,2] = alphaI
    B[1,2,0] = alphaI
    return(B)
def computeGamma(m):
    G = np.zeros(shape=(4,4))
    K = computeK(m)
    for j in range(4):
        for k in range(4):
            if j==k: 
                G[j,k] = sum([m[i]*K[i,j]*(1-K[i,j]) for i in range(4)])
            else:
                G[j,k] = sum([m[i]*K[i,j]*(-K[i,k]) for i in range(4)])
    return(G)
    
def refinedMF(m0,T):
    m = m0
    v = np.zeros((4))
    w = np.zeros((4,4))
    
    M=[]
    V=[]
    W=[]
    
    for t in range(T):
        A = computeA(m)
        B = computeB(m)
        Gamma = computeGamma(m)
        
        M.append(m)
        
        v = np.dot(A,v) + np.tensordot(B,w,2)/2
        w = Gamma + np.dot(np.dot(A,w),np.transpose(A))
        m = np.tensordot(m,computeK(m),1)
        
        V.append(v)
        W.append(w)
        
    return(np.array(M),np.array(V),np.array(W))

M,V,W = refinedMF(m0,500)
```

```python
N=10
plt.plot(M[:,0])
plt.plot(M[:,0]+V[:,0]/N,'--')
plt.grid()
#plt.ylim([0,0.4])
```

# Comparison with simulations 

Note : in order to avoid redoing the 50.000 simulations each time (which takes ~1h), the results are saved in a file. It is possible to force the program to recompute all simulation either by deleting this file or by using "forceRedoSimulation=True". 

```python
import os 
import sys
import time

def simulate_MC(m0,T,N):
    m = np.array([int(N*m0[i]) for i in range(4)])
    #m0
    M = []
    
    for t in range(T):
        M.append(m)
        K = computeK(m*1./N)
        newM = np.zeros(4)
        for i in range(4):
            newM += np.random.multinomial(m[i], K[i,:], size=1).reshape(4)
        m = newM
    return(np.array(M)*1./N)
def averageSimulation(m0,T,N,number_of_simulations,forceRedoSimulation=False):
    fileName = 'traj/simu_SEIR_m{}_{}_{}_{}_N{}_T{}.npz'.format(int(m0[0]*1000),
                                                           int(m0[1]*1000),
                                                           int(m0[2]*1000),
                                                           int(m0[3]*1000),
                                                           N,
                                                           T)
    if os.path.isfile(fileName) and not forceRedoSimulation :
        data = np.load(fileName)
        alreadyDoneSimu = data['nbSimu']
        M = data['M']
    else:
        alreadyDoneSimu = 0
        M = np.zeros(shape=(T,len(m0)))
    if alreadyDoneSimu < number_of_simulations:
        for i in range(alreadyDoneSimu,number_of_simulations):
            M += simulate_MC(m0,T,N)
            print('\r{}/{}'.format(i,number_of_simulations),end='')
        np.savez(fileName,M=M,nbSimu=number_of_simulations)
    return(M/max(alreadyDoneSimu,number_of_simulations))

#m0 = np.array([1,0,0,0])
m0 = np.array([.2,.2,.2,.4])

for N in [10,20,50,100,200,1000]:
    for numberSimu in range(0,(10001 if N<=100 else 1001),1000):
        # This code will generate 50000 simulation and saves the result each 5000 simulation (hence it can be interupted)
        Msimu = averageSimulation(m0,T=500,N=N,number_of_simulations=numberSimu)
```

```python
m0 = np.array([.2,.2,.2,.4])
M,V,W = refinedMF(m0,500)
for N in [10,20,50,100,1000]:
    f = plt.figure()
    Msimu = averageSimulation(m0,T=500,N=N,number_of_simulations=(1000 if N <= 100 else 1000))
    plt.plot(M[:,0])
    plt.plot(Msimu[:,0])
    plt.legend(['$x_S(t)$ (mean field)','$E[X_S[t]]$ (Simulation, $N={}$)'.format(N), '$x_S(t)+V(t)/N$ (refined mean field)'])
    if N==10 or N==1000:
        f.savefig('output_pdfs/SEIR_onlyMF_N{}.pdf'.format(N),bbox_inches='tight')    
    if N<=100:
        plt.plot(M[:,0]+V[:,0]/N,'--')
        plt.legend(['$x_s(t)$ (mean field)','$E[X_S[t]]$ (Simulation, $N={}$)'.format(N), '$S(t)+V(t)/N$ (refined mean field)'])
        f.savefig('output_pdfs/SEIR_all_N{}.pdf'.format(N),bbox_inches='tight')
```
